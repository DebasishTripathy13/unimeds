```python
# ==============================================================================
# Step 1: Install Necessary Packages and Prepare Environment
# ==============================================================================
print("🚀 Setting up the environment...")

# Update and install build tools (g++, make) and git
!sudo apt-get update
!sudo apt-get install -y g++ git make ffmpeg

# Clone the whisper.cpp repository
!git clone https://github.com/ggerganov/whisper.cpp.git

# Change into the repository directory
%cd whisper.cpp

# ==============================================================================
# Step 2: Download the Whisper Model
# ==============================================================================
# We'll use the 'medium' model for good accuracy.
print("Downloading the 'medium' Whisper model...")
!bash ./models/download-ggml-model.sh medium

# ==============================================================================
# Step 3: Build the whisper.cpp Executable
# ==============================================================================
print("Building the whisper.cpp executable...")
!make

# The `make` command creates a suite of executables.
# We'll check for both `main` and `whisper-cli`.
import os

# ==============================================================================
# Step 4: Upload Your Audio File
# ==============================================================================
# This cell is for uploading your audio file. It will prompt a file browser.
print("\n📁 Please upload your Hindi audio file now.")
from google.colab import files
uploaded = files.upload()

# Get the filename of the uploaded file
uploaded_filename = list(uploaded.keys())[0]
print(f"✅ Uploaded '{uploaded_filename}' successfully!")

# ==============================================================================
# Step 5: Run the C++ Transcription and Translation
# ==============================================================================
# This command runs the whisper-cli executable.
# The -l "hi" flag specifies Hindi as the input language.
# The -tr flag tells the model to translate the text to English.
# The -f flag specifies the input audio file.
print("\n--- Starting Transcription and Translation ---")
if os.path.exists("./whisper-cli"):
    print("Using `whisper-cli` executable...")
    !./whisper-cli -m ./models/ggml-medium.bin -l hi -tr -f "{uploaded_filename}"
elif os.path.exists("./main"):
    print("Using `main` executable...")
    !./main -m ./models/ggml-medium.bin -l hi -tr -f "{uploaded_filename}"
else:
    print("Error: Executable not found. Please check the build process.")
```
