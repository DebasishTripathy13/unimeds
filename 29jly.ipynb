```python
import torch
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import soundfile as sf
import numpy as np
import os
import librosa # Added for audio loading

def transcribe_hindi_whisper_finetuned(file_path):
    """
    Uses a Whisper model fine-tuned for Hindi
    """
    # Hindi fine-tuned Whisper model - changed to the user-specified model
    model_name = "collabora/whisper-base-hindi"

    processor = WhisperProcessor.from_pretrained(model_name)
    model = WhisperForConditionalGeneration.from_pretrained(model_name)

    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = model.to(device)

    # Load audio using librosa
    audio, rate = librosa.load(file_path, sr=16000) # Whisper expects 16kHz sample rate

    # Process
    input_features = processor(audio, sampling_rate=rate, return_tensors="pt").input_features.to(device)

    # Generate with Hindi language prompt
    # forced_decoder_ids = processor.get_decoder_prompt_ids(language="hindi", task="transcribe") # Removed this line

    with torch.no_grad():
        predicted_ids = model.generate(
            input_features,
            # forced_decoder_ids=forced_decoder_ids, # Removed this argument
            max_length=448
        )

    transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
    return transcription

# Usage example for fine-tuned model
if __name__ == "__main__":
    audio_path = "/content/X.MP3"

    try:
        print("Using Hindi Fine-tuned Whisper Model...")
        result = transcribe_hindi_whisper_finetuned(audio_path)
        print("Hindi Transcription:")
        print(result)
    except FileNotFoundError as e:
        print(e)
    except Exception as e:
        print(f"Error: {e}")
```

## Load and chunk audio

### Subtask:
Load the audio file and split it into smaller segments.


**Reasoning**:
Load the audio file and split it into smaller segments based on the instructions.



```python
from pydub import AudioSegment

# Define the path to the audio file
audio_path = "/content/X.MP3"

# Load the audio file
audio = AudioSegment.from_file(audio_path)

# Define the desired chunk length in milliseconds (e.g., 30000 for 30 seconds)
chunk_length_ms = 30000

# Initialize an empty list to store the audio chunks
audio_chunks = []

# Iterate through the audio from the beginning to the end, creating chunks
for i in range(0, len(audio), chunk_length_ms):
    chunk = audio[i : i + chunk_length_ms]
    audio_chunks.append(chunk)

# Print the number of chunks created
print(f"Number of chunks created: {len(audio_chunks)}")
```

## Transcribe chunks

### Subtask:
Iterate through the audio chunks and transcribe each one using the Whisper model.


**Reasoning**:
Iterate through the audio chunks, transcribe each chunk using the fine-tuned Whisper model, and store the transcriptions.



```python
import torch
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import soundfile as sf
import numpy as np
import os
import librosa
import tempfile

# Define the model name
model_name = "collabora/whisper-base-hindi"

# Load the processor and model
processor = WhisperProcessor.from_pretrained(model_name)
model = WhisperForConditionalGeneration.from_pretrained(model_name)

# Determine the device
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

# Initialize list to store transcriptions
transcriptions = []

# Iterate through chunks and transcribe
for i, chunk in enumerate(audio_chunks):
    print(f"Processing chunk {i+1}/{len(audio_chunks)}...")
    # Create a temporary file
    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
        temp_file_path = tmp_file.name

    # Export chunk to temporary file
    chunk.export(temp_file_path, format="wav")

    try:
        # Load the temporary audio file using librosa
        audio, rate = librosa.load(temp_file_path, sr=16000)

        # Process the audio
        input_features = processor(audio, sampling_rate=rate, return_tensors="pt").input_features.to(device)

        # Generate transcription
        with torch.no_grad():
            predicted_ids = model.generate(
                input_features,
                max_length=448
            )

        # Decode transcription
        transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
        transcriptions.append(transcription)
        print(f"Chunk {i+1} transcribed.")
    except Exception as e:
        print(f"Error processing chunk {i+1}: {e}")
        transcriptions.append("[Transcription Error]") # Append placeholder for failed chunks
    finally:
        # Remove the temporary file
        os.remove(temp_file_path)

print("\nTranscription of all chunks complete.")
```

## Combine transcriptions

### Subtask:
Concatenate the transcriptions from each chunk to get the complete transcription.


**Reasoning**:
Concatenate the transcribed chunks into a single string.



```python
# Join the transcriptions with a space
full_transcription = " ".join(transcriptions)

# Print the full transcription
print("\n--- Full Hindi Transcription ---")
print(full_transcription)
```
